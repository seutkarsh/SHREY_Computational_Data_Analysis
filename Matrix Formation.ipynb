# Q1 Matrix formation of order 2*3 using for loop.

```python
r=[]
for x in range(2):
    c=[]
    for y in range(3):
        c.insert(y,input())
    r.insert(x,c)
for z in r:
    print(z)
print("Order = 2 * 3 ")
print(r)
```

# Q2 Take value of no. of rows and column form user and create its dynamic size.

```python
row=int(input("Enter no of row HERE= "))
col=int(input("Enter no of column HERE= "))
r=[]
for x in range(row):
    c=[]
    for y in range(col):
        c.insert(y,int(input()))
    r.insert(x,c)
for z in r:
    print(z)
print("Order = ",row,"*",col)
print(r)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[1,0],[0,1]])
c=np.add(a,b)
print(c)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.array([[1,0],[0,1]])
print("A=",a,"\nB=",b)
c=np.dot(a,b)
print("C=A*B")
print("C=",c)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.linalg.inv(a)
print(b)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.transpose(a)
print(b)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b=np.linalg.det(a)
print(b)
```

```python
import numpy as np
a=np.array([[1,2],[3,4]])
b,c=np.linalg.eig(a)
print("Eigen Value=",b,"\nEigen Vector=",c)
```
